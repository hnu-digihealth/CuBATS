```python
import cubats
from cubats.slide_collection.slide_collection import SlideCollection
cubats.__version__
```

Initialize SlideCollection

```python
my_collection = SlideCollection(
    collection_name="Tumor_Set_01",
    src_dir="/path/to/wsi_files",
    dest_dir="/path/to/output_dir",
    ref_slide="/path/to/reference_wsi.tiff",
    path_antigen_profiles="/path/to/threshold_profiles.json"
)
```

Registration of Slides in my_collection

```python
my_collection.register_slides(microregistration=True,crop="overlap")
```

Tumor Segmentation

```python
my_collection.tumor_segmentation(
    model_path = "path/to/model.onnx",
    reference_slide = "path/to/he_wsi.tiff",
    tile_size = (512, 512),
    output_path = None,
    normalization = False,
    inversion = False,
    plot_results = True
)
```

Quantification

```python
my_collection.quantify_all_slides(masking_mode="pixel-level")
```

Pairwise Antigen Co-expression

```python
my_collection.generate_antigen_pair_combinations(masking_mode="pixel-level")
```

Triplet Antigen Co-expression

```python
my_collection.generate_antigen_triplet_combinations(masking_mode="pixel-level")
```
